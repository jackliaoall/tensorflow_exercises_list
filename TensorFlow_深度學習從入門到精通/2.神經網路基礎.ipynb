2.2.1 Sigmoid激活函數

```python
%tensorflow_version 1.x

import tensorflow
print(tensorflow.__version__)
```

```python
import tensorflow as tf

t = tf.constant([[1,2],[2,0]], tf.float32)
result = tf.nn.sigmoid(t)
session = tf.Session()
session.run(result)
```

2.2.2 Tanh激活函數

```python
import tensorflow as tf

t = tf.constant([[1,2],[2,0]], tf.float32)
result = tf.nn.tanh(t)
session = tf.Session()
session.run(result)
```

2.2.3 ReLU激活函數


```python
import tensorflow as tf

t = tf.constant([[1,2],[2,0]], tf.float32)
result = tf.nn.relu(t)
session = tf.Session()
session.run(result)
```
