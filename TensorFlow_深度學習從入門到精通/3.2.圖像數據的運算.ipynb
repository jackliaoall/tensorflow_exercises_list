```python
from PIL import Image

photo = Image.open('./xiongda.jpg')
photo
```

```python
print(photo.size)
photo = photo.resize([128, 128])
print(photo.size)
photo
```

```python
import numpy as np

Im = np.array(photo)
print(Im.shape)
Im[:,:,0]
```

```python
Im = Im/255
print(Im[:,:,0])
```

```python
from matplotlib import pyplot as plt
plt.imshow(Im)
```

```python
Im1 = Im + 0.5
Im2 = 1 - Im
Im3 = 0.5 * Im
Im4 = Im / 0.5

plt.figure()
fig, ax = plt.subplots(1, 4)
fig.set_figwidth(15)
ax[0].imshow(Im1)
ax[1].imshow(Im2)
ax[2].imshow(Im3)
ax[3].imshow(Im4)
```
